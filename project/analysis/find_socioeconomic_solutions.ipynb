```python
#PS1: Do external factors affect the stock market, if so, how much is performance affected?
covid_df = pd.read_csv(path + "covid_df.csv")
covid_df = covid_df.drop(columns = ["Unnamed: 0"])
display(covid_df)

mean_price_change_covid_df = covid_df["Change In Price (%)"].mean()
# print(mean_price_change_covid_df)

non_covid_df = pd.read_csv(path + "non_covid_df.csv")
non_covid_df = non_covid_df.drop(columns = ["Unnamed: 0"])
display(non_covid_df)
mean_price_change_nonCovid_df = non_covid_df["Change In Price (%)"].mean()
# print(mean_price_change_nonCovid_df)

russia_ukraine_df = pd.read_csv(path + "russia_ukraine_df.csv")
russia_ukraine_df = russia_ukraine_df.drop(columns = ["Unnamed: 0"])
display(russia_ukraine_df)

mean_price_change_ru_df = russia_ukraine_df["Change In Price (%)"].mean()
# print(mean_price_change_ru_df)

non_russia_ukraine_df = pd.read_csv(path + "non_russia_ukraine_df.csv")
non_russia_ukraine_df = non_russia_ukraine_df.drop(columns = ["Unnamed: 0"])
display(non_russia_ukraine_df)
mean_price_change_nonRU_df = non_russia_ukraine_df["Change In Price (%)"].mean()
# print(mean_price_change_nonRU_df)
```

```python
print("COVID Change in Price: {:.2f}", mean_price_change_covid_df * 100)
print("NON-COVID Change in Price: {:.2f}", mean_price_change_nonCovid_df * 100)
print("Russia-Ukraine War Change in Price: {:.2f}", mean_price_change_ru_df * 100)
print("Non-Russia-Ukraine War Change in Price: {:.2f}",mean_price_change_nonRU_df * 100)
```

```python
'''
As displayed by the above changes in price, the market is clearly affected by external factors--in this case COVID and/or the Russia-Ukraine War.
In fact, it seems the daily return is affected at least 14 percent in these times.
'''
```
