```python
#Merge market and external indices for analysis
market_external_df = pd.merge(global_market_df, external_indices_df, on = "Date", how = "inner")
display(market_external_df)
```

```python
#One hot encoding
encoded_market_external_df = pd.get_dummies(market_external_df)
display(encoded_market_external_df)
```

```python
cutoff = 1000 #Search endlessly for what could be considered the largest recorded value of CCPI/ECPI ever or of the last 10 years and found nothing. 1000 is arbitrarily chosen and still more than likely incredibly large, but I want to see if I can maintain some data integrity here
encoded_market_external_df = encoded_market_external_df.loc[((encoded_market_external_df["CCPI"] <= cutoff) & (encoded_market_external_df["ECPI"] <= cutoff))]
display(encoded_market_external_df)
```
