```python
import glob
import os
import pandas as pd
```

```python
#Initializing variables and Data
path_to_indices = "C:\\Users\\millen11\\Dropbox\\PC\\Documents\\academia\\rpi\\classes\\fall23\\introToML\\introToMLapps\\project\\dataset\\cleaned\\market_indices"
file_pattern = "*.csv"
file_paths = glob.glob(os.path.join(path_to_indices, file_pattern))

indices = [] #All of the market indices
for file_path in file_paths:
  with open(file_path, "r") as file:
    index_df = pd.read_csv(file_path) #Current market index i.e. S&P500
    indices.append(index_df)
print(len(indices))
```

```python
for market_index in range(len(indices)): #Change unnamed column to Date
  indices[market_index] = indices[market_index].rename(columns = {"Unnamed: 0": "Date"})
# for market in indices:
#   display(market)
```

```python
#Merge every market index
main_market = indices[0]
for market_index in range(1, len(indices) - 1):
  main_market = pd.concat([main_market, indices[market_index]], axis = 0)

display(main_market.loc[main_market["Date"] == "2010-01-04"]) #Example of a day's performance, globally
display(main_market) #Global perfomance of the stock market from 2010 to 2023
```

```python
#One-hot encoding the global peformance for later analysis
main_market_encoded = pd.get_dummies(main_market)
main_market_encoded = main_market_encoded.reset_index(drop = True)
display(main_market_encoded)
```

```python
pre_path = "C:\\Users\\millen11\\Dropbox\\PC\\Documents\\academia\\rpi\\classes\\fall23\\introToML\\introToMLapps\\project\\dataset\\cleaned\\"
main_market_encoded.to_csv(pre_path + "market_indices.csv", index = False)
```
